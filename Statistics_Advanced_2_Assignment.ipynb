**Question 1:** What is hypothesis testing in statistics?

**Answer 1:**
  - Hypothesis testing is a statistical method used to make decisions or inferences about a population based on sample data. It tests an assumption (hypothesis) using probability theory. The process involves:
    1. Formulating a null hypothesis (H₀) and an alternative hypothesis (H₁).
    2. Choosing a significance level (α).
    3. Collecting and analyzing sample data.
    4. Making a decision to reject or fail to reject H₀ based on the test statistic and p-value.


**Question 2:** What is the null hypothesis, and how does it differ from the alternative
hypothesis?

**Answer 2:**
  - Null Hypothesis (H₀): A statement that there is no effect or no difference. It represents the default assumption (e.g., μ = 50).

  - Alternative Hypothesis (H₁ or Ha): A statement that is opposite to H₀, suggesting there is an effect or a difference (e.g., μ ≠ 50).

  - Example: If a company claims the average battery life of a phone is 10 hours:

    - H₀: μ = 10 hours

    - H₁: μ ≠ 10 hours


**Question 3:** Explain the significance level in hypothesis testing and its role in deciding
the outcome of a test.

**Answer 3:**
  - The significance level (α) is the probability of rejecting the null hypothesis when it is actually true.

  - Common values are 0.05 (5%).

  - If the p_value ≤ α → reject H₀ (evidence against H₀).

  - If the p_value > α → fail to reject H₀ (insufficient evidence against H₀).

**Question 4:** What are Type I and Type II errors? Give examples of each.

**Answer 4:**
  - **Type I Error (False Positive):** Rejecting H₀ when it is actually true.
    - Example: A medical test says a healthy person has a disease.

  - **Type II Error (False Negative):** Failing to reject H₀ when it is false.
    - Example: A medical test fails to detect a disease in a sick person.


**Question 5:** What is the difference between a Z-test and a T-test? Explain when to use
each.

**Answer 5:**
  - **Z-test:** Used when the population standard deviation is known and sample size is large (n > 30).

  - **T-test:** Used when the population standard deviation is unknown and sample size is small (n ≤ 30).

  - In practice, the t-test is often used more frequently since population variance is rarely known.

```python
'''
Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]
(Include your Python code and output in the code box below.)
'''

# Answer 7:

   # We test whether the mean of the sample = 50.

import numpy as np
from scipy import stats


sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

data = np.array(sample_data)

# Hypothesized population mean
mu_0 = 50

# Calculate sample mean and std
mean = np.mean(data)
std = np.std(data, ddof=1)
n = len(data)

# Z statistic
z_stat = (mean - mu_0) / (std / np.sqrt(n))

# p-value (two-tailed)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Reject Null Hypothesis (mean ≠ 50)")
else:
    print("Fail to Reject Null Hypothesis (mean = 50)")
```

```python
'''
Question 8: Write a Python script to simulate data from a normal distribution and
calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
(Include your Python code and output in the code box below.)
'''

# Answer 8:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Generate data
data = np.random.normal(loc=100, scale=15, size=200)

# Sample statistics
mean = np.mean(data)
std_err = stats.sem(data)

# 95% confidence interval
conf_interval = stats.t.interval(0.95, len(data)-1, loc=mean, scale=std_err)

print("Sample Mean:", mean)
print("95% Confidence Interval:", conf_interval)

# Plot data
plt.hist(data, bins=20, edgecolor='black', alpha=0.7)
plt.axvline(conf_interval[0], color='red', linestyle='--', label="Lower 95% CI")
plt.axvline(conf_interval[1], color='red', linestyle='--', label="Upper 95% CI")
plt.axvline(mean, color='blue', linestyle='-', label="Sample Mean")
plt.title("Normal Distribution with 95% CI")
plt.legend()
plt.show()
```

```python
'''
Question 9: Write a Python function to calculate the Z-scores from a dataset and
visualize the standardized data using a histogram. Explain what the Z-scores represent
in terms of standard deviations from the mean.
(Include your Python code and output in the code box below.)
'''

# Answer 9:

import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def calculate_z_scores(data):
    mean = np.mean(data)
    std = np.std(data)
    return (data - mean) / std

# Generate dataset
data = np.random.normal(50, 10, 200)
z_scores = calculate_z_scores(data)

# Plot histogram of Z-scores
plt.hist(z_scores, bins=20, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

print("First 10 Z-scores:", z_scores[:10])

'''
Explanation:

A Z-score tells us how many standard deviations a data point is from the mean.

Z = 0 → exactly at the mean

Z = +1 → one std above the mean

Z = -2 → two std below the mean
'''
```
